```python
# sim_euclidean_distance
# Returns a Euclidean distance-based similarity score for person1 and person2
# prefs is the dictionary of persons and their preferences.  For example, critics, above.
# perfect similarity would be 1, Perfect dissimilarity would be 0
def sim_euclidean(prefs, person1, person2):
    debug = False  # if set to True, then we print some extra stuff out while running.
    
    # Get the list of shared_items:
    shared_items = []
    
    # This says: for each movie in the list of movies rated by person1,
    # if that movie is also in the list of movies rated by person2, 
    # then add that movie to shared_items, 
    for item in prefs[person1].keys():  # the keys are the movies.
        if item in prefs[person2].keys():
            shared_items.append(item) 
    
    # if they have no movies in common, return 0:
    if len(shared_items)==0: 
        return 0
    if debug:
        print("shared_items: ", shared_items)
    # Now let's make a list of each of the differences between their ratings of movies they've both seen.
    # For example, [5, 0, -4] would mean that there was a difference of 5 for the first movie in shared_items,
    #                        no difference in their ratings of the second movie in shared_items,
    #                        and a difference of -4 in the ratings of the third movie in shared_items. 
    differences = [ prefs[person1][item] - prefs[person2][item] for item in shared_items ]
    if debug: print("differences: ", differences)
    
    # To calculate the euclidean distance, it's like using the Pythagorean theorem:
    # a-squared + b-squared = c-squared.
    # So now we want the sqaures of those differences (which we'll then add up to get the "c-squared")
    
    differencesSquared = [d*d for d in differences]
    if debug: print("differencesSquared == ", differencesSquared)
    
    # Sum up the squares of all the differences:
    sum_of_squares = sum(differencesSquared)  # the sum function adds up the items in a list
    
    # We also could have done the last couple of steps in one fell swoop:
    #sum_of_squares = sum([pow(prefs[person1][item] - prefs[person2][item], 2) for item in shared_items]  )
    
    # Now, in accordance with the Pythagorean Theorem,
    # we just take the square root of that sum_of_squares to get the overall distance between ratings.
    overall_distance = sqrt(sum_of_squares)
    if debug: print("overall distance == ", overall_distance)
        
    # But we'd like to make this on a scale where the minimal distance (i.e., they have all the same ratings)
    # is a perfect similarity score of 1, whereas if they have a huge difference between their ratings,
    # then there's zero similarity.  So we'll return 1/1+overall_distance.  
    # That way, if overall_distance is 0 (they have the same ratings), then it returns 1/1.
    # But if there's a big difference, then the bigger the overall_difference, the closer the similarity is to 0.
    
    return 1/(1+overall_distance)
```

```python
# sim_pearson
# Returns the Pearson correlation coefficient for p1 and p2
# (The Pearson correlation corrects for "grade inflation", where
#  one critic is consistently harsher than another, but otherwise, their
#  ratings rise and fall together.)
# The Pearson similarity works like this:
# Plot one person's scores on the x axis, and another person's scores on the y axis.
# Then each movie that they both rated can be plotted on the chart, like:
# if Jim gave Superman a 2 and Lisa gave it a 5, then Superman's position is (2,5).
# Do that for all the movies Jim and Lisa have both rated.
# Now try to draw the line of best fit through all those movie points plotted
# on the graph. If the movies all fall exactly on the line, in a diagonal, then there's a 
# perfect correlation between their ratings, and they get a Pearson score of 1.
# If the movies are far from that line, then they get a low Pearson score.
# If there's no correlation, the score is close to zero.
# If there's a negative correlation (Lisa loves the movies Jim hates and vice versa), 
# the score is closer to -1.
# See p. 13

def sim_pearson(prefs, p1, p2): # this one doesn't seem to work.
    # Get the list of mutually rated items
    shared_items = {}
    for item in prefs[p1].keys():
        if item in prefs[p2]: 
            shared_items[item]=1
    
    # Find the number of elements
    n=len(shared_items)
    
    # if they have no ratings in common, return 0
    if n==0: return 0
    
    # Add up all the preferences (ratings) for movies they both rated
    sum1 = sum([prefs[p1][item] for item in shared_items])
    sum2 = sum([prefs[p2][item] for item in shared_items])
    
    # Sum up the squares
    sum1Sq = sum([pow(prefs[p1][item],2) for item in shared_items])
    sum2Sq = sum([pow(prefs[p2][item],2) for item in shared_items])
    
    # Sum up the products
    pSum = sum([prefs[p1][item] * prefs[p2][item] for item in shared_items])
    
    # Calculate Pearson score
    numerator = pSum - (sum1*sum2/n)
    denominator = sqrt((sum1Sq - pow(sum1,2)/n) * (sum2Sq - pow(sum2,2)/n))
    if denominator==0: return 0
    
    r = numerator/denominator
    return r
```

```python
# topMatches
# Returns the best matches for person from the prefs dictionary.
# In other words, which people have the most similar ratings to me?
# Number of results (n) and similarity function are optional params.
# p. 14
def topMatches(prefs, person, n=5, similarity=sim_pearson):
    # Here we'll make a list of the similarity scores
    # for 'person' and each of the other people ('other') 
    scores = [( round(similarity(prefs, person, other), 4), other)
                    for other in prefs if other != person]
    
    # Sort the list so highest scores appear at the top:
    scores.sort()
    scores.reverse()
    return scores[0:n]
```

```python
# getRecommendations
# gets recommendations for a person by using a weighted average
# of every other user's rankings 
# (In other words, if another person has a high similarity score to me,
#  then their reviews get weighted higher -- I should trust them more.) 
# This returns a ranked list of movies with a guess at what my ratings
# for each movie might be
# p. 16
# n is the number of recommendations you want returned.
def getRecommendations(prefs, person, n=10, similarity=sim_pearson):
    if (person not in prefs): 
        print ("Sorry, but I don't see '" + person + "' in my list.  (Check spelling error?)")
        return
    
    #print('Here are the recommendations for ' + person )
    
    # totals is a dictionary of movies, and for each movie there will be a 
    # number representing how much I should see it.
    totals={}
    
    # numOthers is a dictionary of movies, and for each movie there will be a 
    # number that is the number of other people who rated this movie.
    # We'll use this to normalize our rankings.
    # (Otherwise, movies ranked by a lot of people would automatically be more
    # recommended, even if a lot of them ranked it low)
    numOthers={}
    
    # Let's look at each other person in the list of critics ("prefs")
    for other in prefs:
        # don't compare me to myself
        if other==person: continue
        
        # get similarity score between me and this other person.
        sim = similarity(prefs,person,other)
        
        # ignore scores of zero or lower
        if sim <= 0: continue
            
        # Now take a look at each of the movies in this other person's list:
        for item in prefs[other].keys() :         
            # only score movies I haven't seen yet
            if item not in prefs[person].keys() :
                # total  = Similarity * ranking
                totals.setdefault(item,0)  # If this movie isn't in the totals list yet, set it to 0.
                totals[item] += sim * prefs[other][item]  # add this person's rating * their sim score.
                
                # numOthers (We need this for normalizing)
                numOthers.setdefault(item,0) # add the movie if it isn't in there yet.
                numOthers[item] += 1   # add one more person to numOthers
                
    # Create the normalized list,
    # dividing the totals (which are a sum of the rankings * similarities) 
    # by the number of people who reviewed them  :
    rankings = [(round(total/numOthers[item], 4) ,item) for item,total in totals.items()]
    
    # Return the sorted list
    rankings.sort()
    rankings.reverse()
    if len(rankings) < n:
        return rankings
    else:
        return rankings[0:n]
```

```python
# transformPrefs
# This swaps the critics and the movies 
# (The person reviewing with the thing reviewed)
# So that we can, for instance, get similarity ratings on movies 
# (or products, or whatever)
def transformPrefs(prefs):
    result = {}
    for person in prefs:
        for item in prefs[person]:
            result.setdefault(item,{})
            
            # Flip the item and person
            result[item][person] = prefs[person][item]
    return result
```

```python
# ifYouLikeThisMovie
# give me a movie you liked and I'll give you
# a list of other movies you might like
def ifYouLikeThisMovie(movie, n=5, similarity=sim_euclidean):
    movies = transformPrefs(critics)
    if (movie not in movies):
        print ("I don't see '" + movie + "'. (Check for spelling errors?)")
        return
    
    print ('If you liked ' + movie + ' then you might like: ')
    matches = topMatches(movies, movie, n, similarity)
    for item in matches:
        print (item[1], "(", item[0], ")")  # each item in matches is: (similarityScore, movieName)
```
